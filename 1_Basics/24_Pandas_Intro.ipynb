# Pandas: Cleaning

```python
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df.info()
```

```python
df.job_posted_date[0]
```

```python
type(df.job_posted_date[0])
```

```python
pd.to_datetime(df.job_posted_date)
```

```python
df.info()
```

```python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df.info()
```

```python
df.job_posted_month = df.job_posted_date.dt.month
```

```python
df.info()
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
df.info()
```

```python
df
```

```python
df.sort_values(by= 'job_posted_date', inplace=True)
```

```python
df
```

```python
df.drop(labels='salary_hour_avg', axis= 1, inplace= True)
```

```python
df.info()
```

```python
df.dropna(subset=['salary_year_avg'], inplace= True)
```

```python
df.info()
```

# Pandas: Analysis

```python
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.describe()
```

```python
df.info()
```

```python
df.count()
```

```python
df['salary_year_avg'].median()
```

```python
df['salary_year_avg'].min()
```

```python
min_salary = df['salary_year_avg'].idxmin()
```

```python
df.iloc[min_salary]
```

```python
df['job_title_short'].unique()
```

```python
df['job_title_short'].value_counts()
```

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
df.groupby(['job_title_short', 'job_country'])['salary_year_avg'].median()
```

```python
df.groupby('job_title_short')[['salary_year_avg', 'salary_hour_avg']].median()
```

```python
df.groupby('job_title_short')['salary_year_avg'].agg(['min', 'max', 'median'])
```
